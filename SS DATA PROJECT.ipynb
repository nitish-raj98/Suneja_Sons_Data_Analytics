```python
df['Po Placing Date'] = pd.to_datetime(df['Po Placing Date'], format='%d-%m-%Y')
```

```python
print(df.shape)
```

```python
# Replace all spaces in column names with underscores
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace(' ', '')
df.columns = df.columns.str.replace('.', '')
df.columns = df.columns.str.replace('_VD/\nTPC_', 'VD_TPC')
df.columns = df.columns.str.replace('_NSR_', 'NSR')
df.columns = df.columns.str.replace('Mill_/_QSC', 'PLANT')
df.columns = df.columns.str.replace('Qty_\nMt_', 'QTY-MT')
```

```python
print(df.columns)
```

```python
df = df.drop(columns=['X'])
```

```python
print(df.dtypes)
```

```python
# Check total duplicates
print(f"Total duplicate rows: {df.duplicated().sum()}")

# Check duplicates by specific column (like PO_No)
print(f"Duplicate PO_No: {df.duplicated(subset=['PO_No']).sum()}")

# See the duplicate rows
print(df[df.duplicated(subset=['PO_No'], keep=False)])
```

```python
# Get summary statistics
print(df.describe())
```

```python
df.to_csv('SS_DATA_PROJECT_cleaned.csv', index=False)  # Cleaned version

print("✅ Cleaned data saved as 'SS_DATA_PROJECT_cleaned.csv'")
print(f"Total rows: {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
```
